```python
def interpret_line(line):
    decoded_line = line #line.decode('ascii')
    char0 = decoded_line[0]
    if char0 == 'T':
        print("Time: {}".format(decoded_line[2:-1]))
    elif char0 == 'U':
        print("User: {}".format(decoded_line[21:-1]))
    elif char0 == 'W':
        print("Content: {}".format(decoded_line[2:-1]))
```

```python
import gzip
def probe_file(fpath, num_lines):
    with gzip.open(fpath, 'rt') as textf:
        for i, line in enumerate(textf):
            print(line)
            interpret_line(line)
#             print(line[0])
#             print(line.decode('ascii'))
            if i >= num_lines:
                break
```
